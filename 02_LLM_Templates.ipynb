```python
from transformers import LlamaTokenizer, LlamaForCausalLM, GenerationConfig, pipeline
import torch

from langchain.llms import HuggingFacePipeline
```

```python
model_id = "chavinlo/alpaca-native"

tokenizer = LlamaTokenizer.from_pretrained(model_id)

base_model = LlamaForCausalLM.from_pretrained(
    model_id,
    load_in_8bit=True,
    device_map='auto',
)
```

```python
pipe = pipeline(
    "text-generation",
    model=base_model,
    tokenizer=tokenizer,
    max_length=256,
    temperature=0.6,
    top_p=0.95,
    repetition_penalty=1.2
)
```

```python
llm =  HuggingFacePipeline(pipeline= pipe)
```

```python
from langchain.prompts import PromptTemplate, ChatPromptTemplate

string_prompt = PromptTemplate.from_template("explain {subject} like i am 5")
print(string_prompt)
print("----------------------------------------------------------------------------------")
chat_prompt = ChatPromptTemplate.from_template("explain {subject} like i am 5")
print(chat_prompt)
```

```python
string_prompt_message = string_prompt.format_prompt(subject='giza pyramids')
print(string_prompt_message)
print("----------------------------------------------------------------------------------")
chat_prompt_message = chat_prompt.format_prompt(subject='giza pyramids')
print(chat_prompt_message)
```

```python
template_message = """
Tell me {count} advices to {task}
""".strip()

string_prompt = PromptTemplate(
    template=template_message,
    input_variables=['count', 'task']
)

string_prompt_message = string_prompt.format(count='3', task='lose my weight')
```

```python
string_prompt_message
```

```python
# we can save lang chain prompt using load_prompt/save

# try to pass the prompt to llm model
print(
    llm("explain giza pyramids like i am 5")
)
```
